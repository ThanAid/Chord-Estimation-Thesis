```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Data
epochs = list(range(1, 11))  # Epoch numbers
train_accuracy = [0.4234, 0.4316, 0.4315, 0.4314, 0.4311, 0.4308, 0.4306, 0.4303, 0.4301, 0.4296]  # Training accuracy
val_accuracy = [0.4305, 0.4305, 0.4310, 0.4304, 0.4309, 0.4300, 0.4285, 0.4302, 0.4274, 0.4279]  # Validation accuracy

# Plot
plt.figure(figsize=(6, 4))
sns.lineplot(x=epochs, y=train_accuracy, label='Training Accuracy', marker='o')
sns.lineplot(x=epochs, y=val_accuracy, label='Validation Accuracy', marker='o')
plt.title('Accuracy per Epoch')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.xticks(epochs)
plt.yticks([0.42, 0.43, 0.44, 0.45])
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
